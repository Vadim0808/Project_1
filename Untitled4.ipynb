```python
df = pd.read_csv('Steel_industry_data.csv') #First, we load the data and look at it
df.head()
```

```python
# Here, I want to split the date column into two variables because I think it's important to distinguish the time of day, as theoretically,
# it could affect the type of load at the moment of measurement, as well as the month could have an impact (season). Specifically,
# I want to discard the exact date and day since specific days within the same year might confuse the model and make it hard to find patterns
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y %H:%M')
df['time_of_day'] = df['date'].dt.time
df['month'] = df['date'].dt.month
df.head()
```

```python
report = sv.analyze(df) #Now, let's look at the statistical visualization of each variable
report.show_notebook()  #To see the visualization, please follow the link to Google Colab above
```

```python
df['WeekStatus'] = df['WeekStatus'].map({'Weekday': 1, 'Weekend': 0}) #Convert the variable "WeekStatus" to numerical format
df.head()
```

```python
df = pd.get_dummies(df, columns=['Day_of_week', 'month'], drop_first=False, dtype=int) #Convert variables to binary format
df.head()
```

```python
df['time_of_day'] = df['time_of_day'].apply(lambda x: 1 if 8 <= x.hour < 20 else 0) #Split the time of day into morning and night and make it a binary variable
df.head()
```

```python
encoder = OrdinalEncoder(categories=[['Light_Load', 'Medium_Load', 'Maximum_Load']]) #Encode the target variable
df['Load_Type'] = encoder.fit_transform(df[['Load_Type']])
df['Load_Type'] = df['Load_Type'].astype(int)
df.head()
```

```python
X = df.drop(['Load_Type', 'date'], axis=1) #Split the sample into test and training sets
y = df['Load_Type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
#Standardize the numerical variables
features_to_scale = ['Usage_kWh', 'Lagging_Current_Reactive.Power_kVarh', 'Leading_Current_Reactive_Power_kVarh',
                     'CO2(tCO2)', 'Lagging_Current_Power_Factor', 'Leading_Current_Power_Factor', 'NSM']
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[features_to_scale])
X_test_scaled = scaler.transform(X_test[features_to_scale])
X_train[features_to_scale] = X_train_scaled
X_test[features_to_scale] = X_test_scaled
```

```python
#Now, we build a decision tree to efficiently select the best parameters by conducting cross-validation for the F1 score (macro)
dt_model = DecisionTreeClassifier(random_state=42) #In this line of code, I iterated through a multitude of parameters.
param_grid_dt = {                                  #But I entered the best ones for this model into the cross-validation code for faster processing during the check.
    'max_depth': [24],                             #I used such a range: max_depth from 2 to 31,
    'min_samples_split': [2],                      # min_samples_split from 2 to 10,
    'min_samples_leaf': [1],                       # min_samples_leaf from 1 to 5
}
grid_search_dt = GridSearchCV(dt_model, param_grid_dt, cv=11, scoring='f1_macro')
grid_search_dt.fit(X_train, y_train)
print("Best parameters:", grid_search_dt.best_params_)
y_pred_dt = grid_search_dt.best_estimator_.predict(X_test)
test_f1_score_dt = f1_score(y_test, y_pred_dt, average='macro')
print("F1 score on the test set (macro):", test_f1_score_dt)
```

```python
#Create a report on various metrics for the test and training sets and compare them
y_pred_train_dt = grid_search_dt.best_estimator_.predict(X_train)
y_pred_test_dt = grid_search_dt.best_estimator_.predict(X_test)
report_train_dt = classification_report(y_train, y_pred_train_dt)
report_test_dt = classification_report(y_test, y_pred_test_dt)
print("Classification report for the training set:\n", report_train_dt)
print("Classification report for the test set:\n", report_test_dt)
```

```python
#Create a confusion matrix in percentages showing the percentage of errors the model made in each of the three categories,
# as well as the percentage of correct answers, which are highlighted in yellow
y_pred_test_dt = grid_search_dt.best_estimator_.predict(X_test)
conf_mat_test_dt = confusion_matrix(y_test, y_pred_test_dt)
total_samples_per_class_test_dt = conf_mat_test_dt.sum(axis=1)

conf_mat_percentage_test_dt = ((conf_mat_test_dt / total_samples_per_class_test_dt[:, None]) * 100).round(0)
conf_mat_df_test_dt = pd.DataFrame(conf_mat_percentage_test_dt,
                                   index=grid_search_dt.best_estimator_.classes_,
                                   columns=grid_search_dt.best_estimator_.classes_).astype(str) + '%'
def highlight_max(s):
    is_max = s == s.max()
    numbers = s.str.replace('%', '').astype(float)
    is_max = numbers == numbers.max()
    return ['background-color: yellow' if v else '' for v in is_max]

styled_conf_mat_df_test_dt = conf_mat_df_test_dt.style.apply(highlight_max)
styled_conf_mat_df_test_dt
```
